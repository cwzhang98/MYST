```python
import csv
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def load_tsv_data(path):
    n_frames = []
    with open(path, "r") as f:
        reader = csv.DictReader(f, delimiter="\t")
        for d in reader:
            n_frames.append(int(d["n_frames"]))
    return n_frames    
```

```python
train_tsv_path = "/home3/cwzhang/dataset/mustc/en-de/train_st.tsv"
dev_tsv_path = "/home3/cwzhang/dataset/mustc/en-de/dev_st.tsv"
tst_tsv_path = "/home3/cwzhang/dataset/mustc/en-de/tst-COMMON_st.tsv"
n_frames_train =  load_tsv_data(train_tsv_path)
n_frames_dev =  load_tsv_data(train_tsv_path)
n_frames_tst =  load_tsv_data(train_tsv_path)
```
